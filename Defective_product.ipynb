```python
# !pip install ultralytics
from ultralytics import YOLO
import cv2
import os

# Define the path to the trained YOLO model
model_path = '/content/drive/My Drive/Defective_product_cnn/runs/detect/train/weights/best.pt'
# Folder where predictions will be saved
save_dir = '/content/drive/My Drive/Defective_product_cnn/runs/detect/predicted_product'

# Check if the model file exists
if not os.path.exists(model_path):
    print(f"Error: Model file not found at {model_path}")
    print("Please ensure the previous training cell completed successfully and the path is correct.")
else:
    # Load trained YOLO model
    model = YOLO(model_path)  # Change path to your trained model

    # Create save directory if not exists
    os.makedirs(save_dir, exist_ok=True)

    # Test on one image
    # for img in os.listdir('/content/drive/My Drive/Defective_product_cnn/test_img'):
    # img_path = '/content/drive/My Drive/Defective_product_cnn/test_img'  # Change to your test image path
    results = model('/content/drive/My Drive/Defective_product_cnn/test_img', conf=0.25, save=True, project=save_dir, name='predict')  # confidence threshold 25%

    # Display detection result
    for r in results:
        print(r.boxes.data)  # Bounding box info (x,y,w,h,conf,class)
        r.show()  # Opens image with detections
        # r.save(save_dir="output")  # Saves output if needed
```
